## Masking

```python
import numpy as np
import pandas as pd
import savewally as sw
import matplotlib.pyplot as plt
from matplotlib.image import imread
from PIL import Image
```

Import image

```python
image_path = "Stitch_3/{}".format(sw.GetFilepath(col=0, row=0, k=0))
print(image_path)
im = Image.open(image_path)
plt.imshow(im)
```

```python
img_array = np.array(im)
```

```python
img_array.shape
```

```python
mask = img_array != [255,255,255,0]
print(mask)
```

```python
masked = np.all(mask, axis=2)
```

```python
print(np.shape(masked))
im = plt.imshow(masked, vmin=0, vmax=1)
plt.colorbar(im)
plt.show()
```

```python
def non_white(mask_array, level=8):
    image_paths = []
    for i in range(256):
        for j in range(256):
            if mask_array[-i, j]:
                image_path = "Stitch_3/{}".format(sw.GetFilepath(col=i, row=j, k=level))
            else:
                continue
            image_paths.append(image_path)
    return image_paths
```

```python
images = non_white(masked)
```
